```python
import pandas as pd
```

```python
df = pd.read_csv("missing_data.csv")
df.isnull().sum() #to check how many missing values are there in evry column
```

```python
# now filling missing vlaues to the mean of their columns - we use .fillna method
# BMI has 11 mssing NaN places so lets create a new column for that
df['BMI_fillNA'] = df['BMI'].fillna(df['BMI'].mean())

df
```

```python
# renaming the column names
df =  df.rename(columns={'BMI': 'BMI_Index'})
df
```

```python
# changing datatypees od columns

# df['BMI_Index'] = df['BMI_Index'].astype(int) # this might couse error as that column may have missing values

df['BMI_Index'] = df['BMI_Index'].fillna(df['BMI_Index'].mean()).astype(int) 
df
```

```python
# calculating grouped mean
grouped_mean = df.groupby('BMI_Index')['Glucose'].mean()
grouped_mean
```
