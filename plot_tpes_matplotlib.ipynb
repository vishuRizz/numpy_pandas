```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, color='red', linestyle='-', marker='o')
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,16,81,256,625]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1) # (rows, columns, its column number)
plt.plot(x,y1, color='red')
plt.title('plot 1')

plt.subplot(2,2,4) # putting this at 4th position 
plt.plot(x,y2, color='green')
plt.title('plot 2')
```

```python
# Bar Plot
categories = ['a', 'b' , 'c', 'd', 'e']
values = [360,543,54,34,234]
plt.bar(categories, values, color='purple')
```

```python
# Scatter Plot
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.scatter(x,y, marker='x')
```

```python
# Histograms
data = [1,2,3,3,3,3,4,4,5,5,5,5,5,5,5]
plt.hist(data, bins=5, color='orange')
```
